```python
table_name1 = "product"
table_name2 = "customer"
```

```python
from pyspark.sql.functions import *

# Read the erp data
df1 = spark.read.format("csv").option("header","true").load("Files/Assessment Data/ERP_Data/*.csv")
df2 = spark.read.format("csv").option("header","true").load("Files/Assessment Data/CRM_Data/*.csv")
```
